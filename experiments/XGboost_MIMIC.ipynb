```python
import pandas as pd
from skopt.space import Real, Integer
from driver import Driver
```

# Load Cleaned Master MIMIC-IV Dataset

```python
# Cleaned Master MIMIC Data Set
df_master = pd.read_pickle('../MIMIC_IV/df_ckd_master_clean.pkl')
df_base = pd.read_pickle('../MIMIC_IV/df_ckd_base.pkl')
```

# Evaluate XGBoost Pipeline

```python
XGboost_cfg = {
    'tag': 'xgb', # Feature selection model identifier tag
    'response': 'stage_delta',  # Response variable for binary classification and Cox Proportional Hazards; binary CKD stage progression indicator
    'duration': 'duration', # Duration for response variable; days until CKD stage progression or censorship
    'n_novel': 40, # Number of ranked novel features used in supplement to KFRE
    'split': 'valsets',
    'test_size': 0.1, # Ratio of data held out for final XGboost testing
    'val_size': 0.2, # Ratio of data held out for testing in XGboost validation sets
    'n_valsets': 5, # Number of XGboost validation sets constructed from training data
    'n_bayesian': 10, # Number of iterations used in Bayesian hyperparameter optimization
    'n_folds': 5, # CoxPH cross validation folds
    'penalizer': 0.0007, # CoxPH model penalizer
    'auc_space': [
        Integer(5, 20, name='max_depth'),
        Integer(1, 10, name='min_child_weight'),
        Real(0.5, 3.0, name='gamma'),
        Real(0.6, 1.0, name='subsample'),
        Real(0.6, 1.0, name='colsample_bytree'),
        Real(0.01, 0.6, name='colsample_bylevel'),
        Real(0.01, 0.3, name='learning_rate'),
        Integer(50, 150, name='n_estimators')
    ], # Hyperparameter search space used in Bayesian hyperparameter optimization
}

driver = Driver(XGboost_cfg, df_master, df_base)
driver.ClassicalLearningPipe()
```
